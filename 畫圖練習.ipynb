```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.pi
π = np.pi
θ = np.linspace(0, 2*π, 1000)
```

```python
x=[2,4] #head #答對會跑出的圖
y=[-2,0]
plt.plot(x,y,lw=3,c='#FB8500');

x=[4,2.69] #head2
y=[0,0]
plt.plot(x,y,lw=3,c='#FB8500');

x=[0,1.4] #head3
y=[0,0]
plt.plot(x,y,lw=3,c='#FB8500');

x=np.linspace(0,1.763,1000) #head4
y=-0.65*x**2
plt.plot(x,y,lw=3,c='#FB8500');

r=0.3+0.3*np.sin(7*θ) #hair
x=r*np.cos(θ)
y=r*np.sin(θ)
plt.fill(x,y,c='#E0522E');

r=0.4 #eye1
x=2.3+r*np.cos(θ)
y=0.13+r*np.sin(θ)
plt.plot(x,y,lw=3,c='black');

r=0.4 #eye2
x=1.75+r*np.cos(θ)
y=-0.08+r*np.sin(θ)
plt.plot(x,y,lw=3,c='black');

x=np.linspace(2,2.28,1000) #調整比例
y=-0.56*x**2
plt.plot(x,y,lw=3,c='white');

r=0.4 #eye2 color
x=1.75+r*np.cos(θ)
y=-0.08+r*np.sin(θ)
plt.fill(x,y,c='white');

x=np.linspace(2.1,2.57,1000) #mouth
y=-1.4+0.0079*x**2
plt.plot(x,y,lw=3,c='#FB8500');

r=0.15 #eyeball-1
x=2.5+r*np.cos(θ)
y=0.25+r*np.sin(θ)
plt.fill(x,y,lw=4,c='black');

r=0.15 #eyeball-2
x=2+r*np.cos(θ)
y=0.03+r*np.sin(θ)
plt.fill(x,y,lw=4,c='black');

plt.axis('off')
```
